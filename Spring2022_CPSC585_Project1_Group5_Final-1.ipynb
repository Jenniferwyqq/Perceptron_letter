**3. Implement a perceptron to recognize the letter A by producing output +1 for images of the letter A and -1 for images of all other letters.**

**5. Repeat experiment (3) for the rest of the letters, for a total of 26 trained perceptrons. If any perceptron fails to converge, implement the “pocket algorithm” as described at the end of Section 1.2.1.1 of the textbook on p. 8.**

```python
##########################################
#     PART 1 - Train the perceptrons     #
##########################################

TOTAL_LETTERS_CNT = 26
LETTER_IMAGE_SIZE = 256
PART1_TRAINED_MODEL_FILENAME = "part1_model.json"
threshold = 0.03

# Perceptron for letter image classification
class Perceptron(object): 
    def __init__(self, feature_size, alpha=0.3, max_iteration=150):
        self.weights = [random.random() for i in range(feature_size + 1)]
        self.__alpha = alpha
        self.__max_iteration = max_iteration

    # actication function - sigmoid
    def activation_fn(self, x):
        if x >= 0:
            z = math.exp(-x)
            sig = 1 / (1 + z)
            return 1 if sig >= 0.5 else -1
        else:
            z = math.exp(x)
            sig = z / (1 + z)
            return 1 if sig >= 0.5 else -1

    # predict the image with given weights
    def __predict(self, features):
        features_with_bias = list(features)

        # add init bias = 0
        features_with_bias.append(0)

        sum = 0
        # first is bias
        for i in range(len(features_with_bias)):
            w = self.weights[i]
            sum += w * features_with_bias[i]

        prediction = self.activation_fn(sum)
        
        return prediction

    # learning process to find the proper weights to classify a specific letter
    def fit(self, letter_images, target_letter_vector):
        for j in range(self.__max_iteration):
            match_cnt = 0
            for i in range(52):
                y = self.__predict(letter_images[i])
                e = target_letter_vector[i] - y
                if abs(e) > threshold:
                    new_X = [0]         
                    for j in range(256):
                        cur = letter_images[i][j]
                        new_X.append(cur)
                                      
                    tmp = [self.__alpha * float(e) * (float(item)) for item in new_X]                
                    self.weights = [self.weights[k] + tmp[k] for k in range(257)]
                else:
                    match_cnt += 1
            if match_cnt == 52:
                pass
            
            
# train the model with given data set
def train(training_set):
    #Collect all letter images for training
    letter_images = []
    for i in range(len(training_set)):
        letter_images.append(training_set[i][1])
  
    #train the model for all 26 letters
    trained_model = {}
    for i in range(TOTAL_LETTERS_CNT):
        # For idx i at target_letter_vector, 1 means the current letter matrix in TRAINING_SET[i][1]
        # is the target letter, -1 means not the target letter
        target_letter_vector = [-1] * len(training_set)
        target_letter_vector[i] = 1
        target_letter_vector[i + TOTAL_LETTERS_CNT] = 1
    
        #train model for a single letter
        perceptron = Perceptron(feature_size=LETTER_IMAGE_SIZE)
        perceptron.fit(letter_images, target_letter_vector)
        trained_model[chr( ord('A') + i)] = perceptron.weights

    #Save trained model to a local file
    model_file = open(PART1_TRAINED_MODEL_FILENAME, "w")
    json.dump(trained_model, model_file)
    model_file.close()
    print("Model trained. Dumped model at file " + PART1_TRAINED_MODEL_FILENAME)
    
train(TRAINING_SET)
```

```python
##########################################
#     PART 2 - Train the perceptrons     #
##########################################

import numpy as np
from numpy.random import RandomState
from json import JSONEncoder
from matplotlib import pyplot as plt

PART2_TRAINED_MODEL_FILENAME = "part2_model.json"

# Perceptron for letter image classification using Numpy libs
class NP_Perceptron(object): 
    def __init__(self, feature_size, alpha=0.3, max_iteration=18):
        self.weights = np.random.random(size = feature_size+1)
        self.__alpha = alpha
        self.__max_iteration = max_iteration
        
    # actication function - sigmoid
    def activation_fn(self, x):
        if x >= 0:
            z = np.exp(-x)
            sig = 1 / (1 + z)
            return 1 if sig >= 0.5 else -1
        else:
            z = np.exp(x)
            sig = z / (1 + z)
            return 1 if sig >= 0.5 else -1

    # predict the image with given weights
    def __predict(self, features):
        features_with_bias = np.insert(features, 0, 0)      
        z = self.weights.T.dot(features_with_bias)
        prediction = self.activation_fn(z)

        return prediction

    # learning process to find the proper weights to classify a specific letter
    def fit(self, letter_images, target_letter_vector, target_letter=None):  
        error_rate = []
        accuracy = []
        x_axis = []
        for j in range(self.__max_iteration):
            correct_num = []
            for i in range(52):
                y = self.__predict(letter_images[i])
                e = target_letter_vector[i] - y
                if e == 0:
                    correct_num.append(e) 
                    
                x = np.insert(letter_images[i], 0, 0)
                if abs(e) > threshold:                  
                    self.weights = self.weights + self.__alpha * e * x  
            # record the accuracy
            accuracy.append(len(correct_num)/52)
            # record the error rate
            error_rate.append(1-len(correct_num)/52)
            x_axis.append(j)

        if target_letter != None:
            print("Error rate for training perceptron " + target_letter + ":")
            self.plot_error_rate(error_rate, x_axis)
        
    def plot_error_rate(self, error_rate, epoch):
        width_height = (10, 2)
        plt.figure(figsize=width_height)
        plt.xlabel('epoch')
        plt.ylabel('error rate')
        plt.xticks(range(0,18,2))
        plt.plot(epoch, error_rate)
        plt.show()
        
class NumpyArrayEncoder(JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return JSONEncoder.default(self, obj)
    
# train the model with given data set
def train(training_set): 
    #Collect all letter images for training
    letter_images = np.array([training_set[0][1]])
    for i in range(1, training_set.shape[0]):
        # , training_set[i][1].shape
        letter_images = np.concatenate((letter_images, np.array([training_set[i][1]])), axis=0)
    # print(letter_images)
    #train the model for all 26 letters
    
    trained_model = np.empty([26, 257])
    for i in range(TOTAL_LETTERS_CNT):
        # For idx i at target_letter_vector, 1 means the current letter matrix in TRAINING_SET[i][1]
        # is the target letter, -1 means not the target letter
        target_letter_vector = np.full((training_set.shape[0]), -1, dtype=int)  
        target_letter_vector[i] = 1
        target_letter_vector[i + TOTAL_LETTERS_CNT] = 1
        
        #train model for a single letter
        perceptron = NP_Perceptron(feature_size=LETTER_IMAGE_SIZE)
        perceptron.fit(letter_images, target_letter_vector, chr( ord('A') + i))      
        trained_model[i] = perceptron.weights

    # Serialization
    print("Model trained. Dumped model at file " + PART2_TRAINED_MODEL_FILENAME)
    with open(PART2_TRAINED_MODEL_FILENAME, "w") as write_file:
        json.dump(trained_model, write_file, cls=NumpyArrayEncoder)

    
```
